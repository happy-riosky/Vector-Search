## Tesseract


```python
import pytesseract
from PIL import Image
import cv2
import numpy as np

# 设置 Tesseract 可执行文件的路径（Windows 系统需要）
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def preprocess_image(image_path):
    # 读取图像
    img = cv2.imread(image_path)
    
    # 转换为灰度图像
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # 应用阈值处理以增强对比度
    _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    return thresh

def ocr_from_image(image_path):
    try:
        # 预处理图像
        processed_image = preprocess_image(image_path)
        
        # 使用 PIL 打开处理后的图像
        img = Image.fromarray(processed_image)
        
        # 使用 Tesseract 进行 OCR 识别
        text = pytesseract.image_to_string(img, lang='chi_sim')  # 可根据需要更改语言，如 'chi_sim' 用于中文
        
        return text
    except Exception as e:
        return f"OCR 处理出错: {str(e)}"

# 主函数
if __name__ == "__main__":
    # 替换为你的图像文件路径
    image_path = "/Users/riosky/Pictures/ocr/para.jpeg"
    
    # 执行 OCR
    result = ocr_from_image(image_path)
    
    # 打印识别结果
    print("识别的文本内容：")
    print(result)
```

## PaddleOCR


```python
from paddleocr import PaddleOCR

# 初始化 PaddleOCR 实例
ocr = PaddleOCR(
    use_textline_orientation=True,  # 启用角度分类（处理旋转文本）
    lang='ch',           # 设置语言为中文（支持 80+ 种语言，可改为 'en' 等）
)

# 指定图片路径（可以是本地路径或 URL）
img_path = '/Users/riosky/Pictures/ocr/para.jpeg'  # 替换为你的图片路径

# 执行 OCR 推理
result = ocr.predict(img_path)

# 输出识别结果
for line in result:
    box = line[0]           # 文字区域的边界框坐标
    text, score = line[1]   # 识别的文本和置信度
    print(f"Text: {text}, Confidence: {score}, Box: {box}")
```
